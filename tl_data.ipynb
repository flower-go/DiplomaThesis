```python
cesta = "/Users/petravysusilova/Documents/school/diplomka_main/diplomka_text/diplomka/"
name = "tl_1_vystup"
dataset = []
i = 0
with open(cesta+name, "r") as file:
    while True:
        #print(i)
        i  =i+1
        line = file.readline()
        if len(line) == 0:
            break
        s = line.split("\t")
        if len(s) > 5: #TODO zeptat se na radku 749
            word, t1, lemma, t2, tag, lemma_true, tag_true, *analysis = line.split("\t")
            dataset.append([word,t1, lemma, t2, tag,lemma_true, tag_true, analysis])
```

```python
df = pd.DataFrame(dataset, columns=["word","lemma_T","lemma","tag_T","tag", "lemma_true","tag_true","analysis"])
```
